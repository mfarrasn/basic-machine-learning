```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target
columns = iris.feature_names

#Dataframe
df_iris = pd.DataFrame(iris.data)
df_iris.columns = columns
df_iris.head()
```

```python
sns.boxplot(x=df_iris['sepal length (cm)'])
```

```python
sns.boxplot(x=df_iris['sepal width (cm)'])
```

```python
Q1 = df_iris.quantile(0.25)
Q3 = df_iris.quantile(0.75)
print(Q1)
print(Q3)
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
outlier = (df_iris < (Q1-1.5*IQR)) | (df_iris > (Q3+1.5*IQR))
pd.set_option('display.max_rows', outlier.shape[0]+1)
outlier
```
