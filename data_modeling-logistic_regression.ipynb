## Preprocess Data

```python
df_banking.head()
```

### Process Categorical Data

```python
list_vars = ['job','marital','education','default','housing','loan','contact','month','day_of_week','poutcome']
for var in list_vars:
    list_vars ='var'+'_'+var
    list_vars = pd.get_dummies(df_banking[var], prefix=var)
    data1 = df_banking.join(list_vars)
    df_banking = data1
```

```python
df_banking.head()
```

```python
df_banking.columns.values
```

### Remove Categorical Data

```python
list_vars = ['job','marital','education','default','housing','loan','contact','month','day_of_week','poutcome']
data_vars = df_banking.columns.values.tolist()
to_keep = [i for i in data_vars if i not in list_vars]
```

```python
df_banking_new = df_banking[to_keep]
df_banking_new.columns.values
```

## Preapare Features & Labels

```python
data_vars = df_banking_new.columns.values.tolist()
y = ['y']
X = [i for i in data_vars if i not in y]
```

## Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_banking_new[X], df_banking_new[y], test_size=0.3, random_state=0)
```

```python
from sklearn import preprocessing
```

```python
X_train = preprocessing.scale(X_train)
X_test = preprocessing.scale(X_test)
```

## Model Data

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

## Test Model

```python
model_score = logreg.score(X_test, y_test)
```

```python
print('Model Accuracy:', model_score)
```
